```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

url = "https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8#%D0%9D%D0%B0%D1%80%D0%BE%D0%B4%D0%B6%D1%83%D0%B2%D0%B0%D0%BD%D1%96%D1%81%D1%82%D1%8C"

df = pd.read_html(url, header=0)[12]
df.head()
```

Видаліть з таблиці дані по всій країні, останній рядок таблиці

```python
df = df.drop(df.index[-1])
df
```

Замініть відсутні дані в стовпцях середніми значеннями цих стовпців (метод fillna)

```python
numeric_cols = df.select_dtypes(include=[np.number]).columns

df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean())
df = df.round(2)
df
```

```python
# Відбір даних для регіону Крим
crimea_data = df[df['Регіон'] == 'Крим']

# Побудова графіка
plt.figure(figsize=(10, 6))
plt.plot(crimea_data.columns[1:], crimea_data.values[0][1:], marker='o')
plt.title('Рівень народжуваності в регіоні Крим за роки')
plt.xlabel('Рік')
plt.ylabel('Рівень народжуваності')
plt.grid(True)
plt.show()
```

Отримайте список регіонів, де рівень народжуваності у 2019 році був вищим за середній по Україні

```python
# Відбір даних для регіону Крим
kyiv_data = df[df['Регіон'] == 'Київ']

# Побудова графіка
plt.figure(figsize=(10, 6))
plt.plot(kyiv_data.columns[1:], kyiv_data.values[0][1:], marker='o')
plt.title('Рівень народжуваності в регіоні Київ за роки')
plt.xlabel('Рік')
plt.ylabel('Рівень народжуваності')
plt.grid(True)
plt.show()
```

```python
# Перетворення значень у числовий тип
df['2019'] = pd.to_numeric(df['2019'], errors='coerce')

# Обчислення середнього значення рівня народжуваності по Україні
average_birth_rate_ukraine = df['2019'].mean()

# Відфільтруємо регіони, де рівень народжуваності у 2019 році був вищим за середній по Україні
regions_above_average = df[df['2019'] > average_birth_rate_ukraine]['Регіон'].unique()


print("Регіони з середнім значення рівня народжуваності по Україні у 2019 році:", average_birth_rate_ukraine)

```

```python
regions_with_higher_birth_rate = df[df['2019'] > average_birth_rate_ukraine]['Регіон']
print(regions_with_higher_birth_rate)
```

```python
filtered_df = df[df['2019'] > average_birth_rate_ukraine]

# Відбір унікальних значень регіонів та відповідних значень народжуваності
unique_regions = filtered_df['Регіон'].unique()
birth_rates = [filtered_df[filtered_df['Регіон'] == region]['2019'].iloc[0] for region in unique_regions]

# Побудова кругової діаграми
plt.figure(figsize=(7, 5))
plt.pie(birth_rates, labels=unique_regions, autopct='%1.1f%%')
plt.title('Регіони з рівнем народжуваності у 2019 році вищим за середній по Україні')
plt.show()
```

У якому регіоні була найвища народжуваність у 2014 році?

```python
# Перетворення значень у числовий тип
df['2014'] = pd.to_numeric(df['2014'], errors='coerce')

# Знайдемо регіон з найвищим рівнем народжуваності у 2014 році
region_with_max_birth_rate_2014 = df.loc[df['2014'].idxmax()]['Регіон']

print("Регіон з найвищим рівнем народжуваності у 2014 році:", region_with_max_birth_rate_2014)
```

Побудуйте стовпчикову діаграму народжуваності по регіонах у 2019 році


```python
# Вибір даних для побудови
data = df[['Регіон', '2019']].copy()
data.dropna(inplace=True)  # Видалення рядків з відсутніми значеннями

# Побудова стовпчикової діаграми
plt.figure(figsize=(10, 5))
plt.bar(data['Регіон'], data['2019'], color='skyblue')
plt.xlabel('Регіон')
plt.ylabel('Народжуваність у 2019 році')
plt.title('Народжуваність по регіонах у 2019 році')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
