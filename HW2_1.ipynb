Прочитайте дані за допомогою методу read_html з таблиці "Коефіцієнт народжуваності в регіонах України (1950—2019)" посилання   Вивести перші рядки таблиці за допомогою методу head

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

url = "https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8#%D0%9D%D0%B0%D1%80%D0%BE%D0%B4%D0%B6%D1%83%D0%B2%D0%B0%D0%BD%D1%96%D1%81%D1%82%D1%8C"

df = pd.read_html(url, header=0)[12]
df.head()
```

Визначте кількість рядків та стовпців у датафреймі (атрибут shape)

```python
num_rows, num_columns = df.shape

print(f"Кількість рядків: {num_rows}")
print(f"Кількість стовпців: {num_columns}")
```

Замініть у таблиці значення "—" на значення NaN

```python
df.replace(["—"], np.nan, inplace = True)
df
```

Визначте типи всіх стовпців за допомогою dataframe.dtypes

```python
df.dtypes
```

Замініть типи нечислових колонок на числові. Підказка - це колонки, де знаходився символ "—"

```python
numeric_cols = df.columns[1:9].astype(float)

print(df)
print(df.dtypes)
```

Видаліть з таблиці дані по всій країні, останній рядок таблиці

```python
df = df.drop(df.index[-1])
df
```

Замініть відсутні дані в стовпцях середніми значеннями цих стовпців (метод fillna)

```python
numeric_cols = df.select_dtypes(include=[np.number]).columns

df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean())
df = df.round(2)
df
```

Отримайте список регіонів, де рівень народжуваності у 2019 році був вищим за середній по Україні

```python
average_birth_rate_ukraine = df['2019'].mean()
print(average_birth_rate_ukraine)
```

```python
regions_with_higher_birth_rate = df[df['2019'] > average_birth_rate_ukraine]['Регіон']
print(regions_with_higher_birth_rate)
```

У якому регіоні була найвища народжуваність у 2014 році?

```python
region_with_highest_birth_rate_2014 = df.loc[df['2014'].idxmax(), 'Регіон']
print(f'Регіон у якому була найвища народжуваність у 2014 році {region_with_highest_birth_rate_2014}')
```

Побудуйте стовпчикову діаграму народжуваності по регіонах у 2019 році


```python
# Вибір даних для побудови
data = df[['Регіон', '2019']].copy()
data.dropna(inplace=True)  # Видалення рядків з відсутніми значеннями

# Побудова стовпчикової діаграми
plt.figure(figsize=(10, 5))
plt.bar(data['Регіон'], data['2019'], color='skyblue')
plt.xlabel('Регіон')
plt.ylabel('Народжуваність у 2019 році')
plt.title('Народжуваність по регіонах у 2019 році')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
